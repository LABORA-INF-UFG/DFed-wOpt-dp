```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
```

```python
def graph_values_acc(uploads_mean_acm, acc_mean, init=0, k=10):
    lista_base = list(range(init, 5001, k))
    indices_maior_ou_igual = [next((i for i, valor in enumerate(uploads_mean_acm) if valor >= x), None) for x in lista_base]
    nova_lista = [valor for valor in indices_maior_ou_igual if valor is not None]
    y = np.array(acc_mean)[nova_lista]
    x = lista_base[:len(y)]
    return x, y
```
